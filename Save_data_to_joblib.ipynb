## Load data and save it as .joblib file

```python
from PIL import Image
import os
import joblib
from skimage.io import imread
from skimage.transform import resize

cwd = os.getcwd()

data_path = os.path.join(cwd,'data')

output = os.path.join(cwd,'resized_data')

def resize_all(src, pklname, width=32, height=32):
    '''
     resize all the image in database
    '''
    height = height if height is not None else width
     
    data = dict()
    data['description'] = f'resized ({width},{height}) animal images in rgb'
    data['label'] = []
    data['filename'] = []
    data['data'] = []   
     
    pklname = f"{pklname}_{width}x{height}px.joblib"
 
    # read all images in PATH, resize and write to DESTINATION_PATH
    for subdir in os.listdir(src):
        if subdir[0]!=".":
            print(subdir)
            current_path = os.path.join(src, subdir)

            for file in os.listdir(current_path):
                if file[-3:] in {'jpg', 'png'}:             
                    im = imread(os.path.join(current_path, file))
                    if im.size!=(32,32):
                        im = resize(im, (width, height)) #[:,:,::-1]
                    data['label'].append(subdir)
                    data['filename'].append(file)
                    data['data'].append(im)
 
        joblib.dump(data, pklname)
    print(f'Store the data at "{pklname}"')

 
resize_all(src=data_path, pklname=output)

```
